# 2. Preprocessing: Pembagian Dataset (Dataset Splitting)
Tahap ini bertujuan untuk membagi dataset mentah (`raw`) menjadi tiga bagian ke dalam folder `processed`:
- **Data Latih (Train): 70%** > Untuk melatih AI (ResNet-50).
- **Data Validasi (Validation): 15%** > Untuk mengevaluasi AI saat proses belajar berlangsung agar tidak menghafal (*overfitting*).
- **Data Uji (Test): 15%** > Untuk ujian akhir AI setelah selesai belajar (blind test).

```python
import splitfolders

input_folder = "../dataset/raw" 
output_folder = "../dataset/processed" 

print("Memulai proses pemisahan data...")

# Membagi dataset dengan rasio 70% Train, 15% Val, 15% Test
# Parameter seed=42 memastikan pembagiannya acak tapi konsisten
splitfolders.ratio(
    input_folder, 
    output=output_folder, 
    seed=42, 
    ratio=(0.7, 0.15, 0.15), 
    group_prefix=None
)

print("Proses pemisahan data selesai! Silakan cek folder dataset/processed/")
```

```python
import os

output_folder = "../dataset/processed" 

print("=== Rangkuman Hasil Splitting ===")

for split in ['train', 'val', 'test']:
    print(f"\nFolder {split.upper()}:")
    split_dir = os.path.join(output_folder, split)
    
    total_split = 0
    for cls in ['COVID', 'Normal', 'Viral_Pneumonia']:
        cls_dir = os.path.join(split_dir, cls)
        
        if os.path.exists(cls_dir):
            count = len([f for f in os.listdir(cls_dir) if f.endswith('.png')])
            print(f"- {cls}: {count} gambar")
            total_split += count
            
    print(f"Total di {split}: {total_split} gambar")
```
