# 1. Eksplorasi Dataset (Exploratory Data Analysis)
Tujuan dari tahap ini adalah untuk:
1. Mengetahui distribusi jumlah citra (gambar) pada masing-masing kelas penyakit.
2. Memvisualisasikan perbandingan jumlah data untuk melihat apakah data seimbang (*imbalanced*).
3. Menampilkan contoh sampel gambar dari tiap kelas (Normal, COVID-19, dan Viral Pneumonia).

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

base_dir = '../dataset/raw'

classes = ['COVID', 'Normal', 'Viral_Pneumonia']

data_counts = {}

for cls in classes:
    folder_path = os.path.join(base_dir, cls)
    images = [f for f in os.listdir(folder_path) if f.endswith('.png')]
    data_counts[cls] = len(images)

# Menampilkan hasil perhitungan
print("=== Distribusi Dataset X-Ray ===")
for cls, count in data_counts.items():
    print(f"- {cls}: {count} citra")

print(f"\nTotal keseluruhan: {sum(data_counts.values())} citra")
```

```python
# (Bar Chart) untuk visualisasi jumlah data
plt.figure(figsize=(8, 5))
plt.bar(data_counts.keys(), data_counts.values(), color=['#ef4444', '#22c55e', '#3b82f6'])

plt.title('Distribusi Jumlah Gambar per Kelas Penyakit', fontsize=14)
plt.xlabel('Kelas Diagnosis', fontsize=12)
plt.ylabel('Jumlah Gambar (Citra)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(data_counts.values()):
    plt.text(i, v + 50, str(v), ha='center', fontweight='bold')

plt.show()
```

```python
# Menampilkan 1 contoh gambar dari masing-masing kelas
plt.figure(figsize=(15, 5))

for i, cls in enumerate(classes):
    folder_path = os.path.join(base_dir, cls)
    sample_img_name = [f for f in os.listdir(folder_path) if f.endswith('.png')][0]
    img_path = os.path.join(folder_path, sample_img_name)
    
    img = mpimg.imread(img_path)
    
    # Tampilkan gambar berdampingan (1 baris, 3 kolom)
    plt.subplot(1, 3, i + 1)
    plt.imshow(img, cmap='gray') # Pakai cmap='gray' karena ini foto rontgen (hitam putih)
    plt.title(f"Kelas: {cls}\nUkuran: {img.shape}", fontsize=12)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
